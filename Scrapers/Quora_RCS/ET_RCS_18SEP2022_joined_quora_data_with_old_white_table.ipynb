```python
import pandas as pd
import numpy as np
import os
```

```python
#helper functions
def if_words_in_tech_list(li, tech_dict):
    shared_words = []
    for names in tech_dict.values():
        if isinstance(names, str):
            names = [names]
        intersect_words  = [word for word in names if word in li]
        
        shared_words.extend(intersect_words)
    return shared_words

def remove_non_letters(text_words):    
    new_text_words = [word for word in text_words if word.isalpha() ] 
    return new_text_words
```

```python
#We're interested in Emerging Tech column for now
tech_values = techs_df[['Emerging Tech', 'wiki_aliases']]
```

```python
tech_values
```

```python
#merge names into tech_list
tech_dict = {}

for i, row in tech_values.iterrows():
    official_name = row['Emerging Tech']
    if isinstance(row['wiki_aliases'], float):
        if np.isnan(row['wiki_aliases']):
            tech_dict[i] = official_name
        continue
    synonyms = row['wiki_aliases'].split(',')
    synonyms.extend([official_name])
    tech_dict[i] = synonyms
```

```python
answer_text = quora_df['Answer'].to_list()
```

```python
#iterate through answers and check if contains a tech in tech_list
ans_tech_dict = {}

for i, ans in enumerate(answer_text):
    #split ans post to words
    ans_words = ans.split(' ') 
    ans_words = remove_non_letters(ans_words)
    
    ans_tech_dict[str(i)] = if_words_in_tech_list(ans_words, tech_dict)
    
    
```
