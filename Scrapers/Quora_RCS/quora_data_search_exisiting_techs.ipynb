```python
#helper functions
def unpack_list(li):
    unpacked_list = []
    for el in li:
        if isinstance(el, list):
            if el == []:
                unpacked_list.extend([np.nan])
            else:
                unpacked_list.extend([*el])
        elif isinstance(el, str):
            unpacked_list.extend([el])
    return unpacked_list
```

```python
quora_table = pd.read_csv(quora_path)
quora_table.drop('Unnamed: 0', axis=1, inplace=True)
quora_table
```

```python
et_table = pd.read_csv(et_table_path)  
```

```python
et_table
```

```python
#keep only synonyms columns (not german aliases)
cols = ['Emerging Tech', 'wiki_aliases']

tech_syn_table = et_table[cols]
```

```python
tech_syn_table
```

```python
#create a dict with Emerging Tech as keys and Emerging Tech and wiki_aliases as values
tech_dict = {}
for i, row in tech_syn_table.iterrows():
    tech_vals = [row['Emerging Tech']]
    if row['wiki_aliases'] is not np.nan:
        tech_vals.extend(row['wiki_aliases'].split(','))
    else:
        pass
    tech_dict[row['Emerging Tech']] = tech_vals
```

```python
tech_dict
```

```python
quora_table
```

```python


mentioned_techs = []


for i, ans_text in enumerate(quora_table['Answer']):
    if ans_text is np.nan:
        mentioned_techs.append([])
      
        continue
    
    existing_techs = []
    for key in tech_dict.keys():
        words = list(set(ans_text.split(' ')))
        ans_techs = [word for word in words if word in tech_dict[key]]
        if ans_techs == []:
            continue
        else:
            existing_techs.extend(ans_techs)

   
    mentioned_techs.append(existing_techs)
    
```

```python
mentioned_techs = unpack_list(mentioned_techs)
```

```python
mentioned_techs = pd.DataFrame(mentioned_techs)
```

```python
quora_table['mentioned techs'] = mentioned_techs  
```
