```python
import pandas as pd
import numpy as np
import ast
import os
from collections import Counter
```

```python
src_path = r"C:\Users\Ravit\Documents\rnd\horizon_scanning_lab\Scrapers\Quora_RCS\answers_and_info"
```

```python
def file_to_dict(file_path):
    with open(file_path) as f:
        lines = f.readlines()
    return ast.literal_eval(lines[0])
```

```python
#bertopic labels
labels_path = os.path.join(src_path, "topic_rep_labels")

#bertopic representations
represent_path = os.path.join(src_path, "topic_representations")
```

```python

labels_dict = file_to_dict(labels_path)
rep_dict = file_to_dict(represent_path)
```

```python
len(rep_dict)
```

```python
labels_dict
```

```python
rep_dict
```

```python
#keep only words in vocabulary
lists_of_words = []
all_words = []

for key in rep_dict.keys():
    words = [li[0] for li in rep_dict[key]]
    lists_of_words.append(words)
    all_words.extend(words)
```

```python
all_words
```

```python
#remove words that appear in two or more lists
count_words = Counter(all_words).most_common()
```

```python
count_words
```

```python
words_to_remove = []
words_to_inspect =[]
for c in count_words:
    if c[1] > 2:
        words_to_remove.append(c[0])
    elif c[1]>1:
        words_to_inspect.append(c[0])
```

```python
words_to_remove
```

```python
words_to_inspect
```

```python
# keep ai, home, cars, business
words_to_remove.extend(words_to_inspect[:7])
```

```python
words_to_remove
```

```python
#add 'using', 'used' to vocabulary
words_to_remove.extend(['using', 'used'])
```

```python
#remove words from vocabulary
out_dict = {}

for i, key in enumerate(rep_dict.keys()):
    keep_words = [li[0] for li in rep_dict[key] if li[0] not in words_to_remove]
    sentence = ''
    for word in keep_words:
        sentence += word + ' '
    out_dict[key] = sentence[:-1]
```

```python
out_dict
```

```python
out_df = pd.DataFrame.from_dict(out_dict, orient='index')
```

```python
out_df
```

```python
out_df.to_csv(os.path.join(src_path, 'words_in_clusters_after_preliminary_cleaning.csv'))
```
