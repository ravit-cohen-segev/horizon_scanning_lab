```python
def file_to_dict(file_path):
    with open(file_path) as f:
        lines = f.readlines()
    return ast.literal_eval(lines[0])
```

```python
#bertopic labels
answers_labels_path = os.path.join(src_path, "answers_topic_rep_labels_second_iteration")


#bertopic representations
answers_represent_path = os.path.join(src_path, "answers_topic_representations_second_iteration")
```

```python
#answers dicts
answers_labels_dict = file_to_dict(answers_labels_path)
answers_rep_dict = file_to_dict(answers_represent_path)

```

```python
#keep only words in vocabulary
def words_in_dict(d): 
    lists_of_words = []
    all_words = []

    for key in d.keys():
        words = [li[0] for li in d[key]]
        lists_of_words.append(words)
        all_words.extend(words)
    return all_words
```

```python
all_words = words_in_dict(answers_rep_dict)
```

```python
#remove words that appear in two or more lists
count_words = Counter(all_words).most_common()
```

```python
count_words
```

```python
set(all_words)
```

```python
#remove words
#remove numbers
def numbers_to_drop_from_list(full_list):
    words_to_remove = []
    for word in full_list:
        for w in word:
            if w.isdigit():
                words_to_remove.append(word)
    return words_to_remove
```

```python
words_to_remove = numbers_to_drop_from_list(all_words)
```

```python
all_words = list(set(all_words) - set(words_to_remove))
```

```python
#remove words from vocabulary
out_dict = {}

for i, key in enumerate(answers_rep_dict.keys()):
    keep_words = [li[0] for li in answers_rep_dict[key] if li[0] in all_words]
    sentence = ''
    for word in keep_words:
        sentence += word + ' '
    out_dict[key] = sentence[:-1]
```
