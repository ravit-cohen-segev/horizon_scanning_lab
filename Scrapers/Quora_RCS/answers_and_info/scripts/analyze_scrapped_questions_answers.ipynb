```python
import nltk

nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
#tokenize words
words_in_quote = word_tokenize(questions_list[1].strip('https://www.quora.com/').replace('-',' '))
```

```python
#remove stop words
def remove_stopwords_from_sentence(sentence):   
    stop_words = set(stopwords.words("english"))
    filtered_list = []
    #tokenize sentence
    words_in_sentence = word_tokenize(sentence.strip('https://www.quora.com/').replace('-',' '))
    for word in words_in_sentence:
       if word.casefold() not in stop_words:
            filtered_list.append(word)
    
    new_sentence = ""
    for worf in filtered_list:
        new_sentence = new_sentence + " " + worf
    return new_sentence


```

```python
import spacy
from spacy.lang.en import English

parser = English()
nlp=spacy.load("en_core_web_sm")
sentence_topics = []
for quest in questions_list:
    doc = nlp(remove_stopwords_from_sentence(quest))
    
    sub_toks = [tok for tok in doc if (tok.dep_ == "nsubj") ]
    sentence_topics.append(sub_toks)
```

```python
sentence_topics
```

```python
sentence_str_topics = []

for i,topic in enumerate(sentence_topics):
    if (topic == []) or (topic is None):
        str_topic = []        
    else:         
        str_topic = [str(w) for w in topic]
    
    if '%' in str_topic:
        str_topic.remove('%')
    if 'internet' in str_topic:
        print(i, questions_list[i])
    sentence_str_topics.append(str_topic)
```

```python
tech_syns = ["technology", "technologies", "advancement", "progress", "peak", "evolve", "evolvement", "develop", "development"]
```

```python


def assign_domain_category(domain_category, syns_words, category):
    
    for sentence in sentence_str_topics:
       if sentence is None or sentence == []:
          domain_category.append([])
          continue
       else:
          for tech in syns_words:
                if tech in sentence:
                    domain_category.append(category)
                    continue
    return domain_category

domain_category = assign_domain_category([], tech_syns, "Tech general")
    
```
