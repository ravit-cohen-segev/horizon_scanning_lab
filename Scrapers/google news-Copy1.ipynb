```python
import pandas as pd
from pygooglenews import GoogleNews
```

```python
gn = GoogleNews()
```

```python
def get_google_news(q,start,end):
    news=gn.search(q, helper = True, from_ =start , to_ =end)
    df=pd.DataFrame(news["entries"])
    return df
    
    

q="boaz"
start="2020-01-01"
end="2020-12-31"
get_google_news(q,start,end)
```
