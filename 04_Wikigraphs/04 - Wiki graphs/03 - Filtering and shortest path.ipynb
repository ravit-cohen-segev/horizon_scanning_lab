```python
import pandas as pd
import sys
import tqdm
from IPython.display import clear_output
import networkx as nx
import pickle
```

```python
#load file in chunks and filter neccery columns
filename="enwiki-20211031-categorylinks.csv"
chunksize = 10 ** 6
i=0
lst2=[]
with pd.read_csv(filename, chunksize=chunksize) as reader:
    for chunk in reader:
#         display(chunk)
        dff2=chunk[["cl_from","cl_to","cl_type"]]
        clear_output(wait=True)
        print(i,dff2.shape)
        i+=1
        if dff2.shape[0]>0:
            lst2.append(dff2.copy())
```

```python
#combine chunks
dff2=pd.concat(lst2)
dff2
```
